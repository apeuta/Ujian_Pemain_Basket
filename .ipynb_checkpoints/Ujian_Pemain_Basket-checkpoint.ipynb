# UJIAN PEMAIN BASKET

```python
import pandas as pd
import numpy as np
```

#### Preparing Data

```python
enbie = pd.read_csv("NBA.csv")
enbie.head()
```

```python
dfRekrut = pd.DataFrame({
    "Name" : ["Andakara Prastawa Dyaksa", "Reggie Mononimbar", "Hardianus Lakudu", 
              "Kevin Yonas Sitorus", "Arki Dikania Wisnu", "Laurentius Steven Oei", 
              "Mei Joni", "Vincent Rivaldi", "Hardian Wicaksono", "Brandon Jawato"], 
    "Club" : ["Pelita Jaya Bakrie", "Pelita Jaya Bakrie", "Satria Muda Pertamina Jakarta", 
              "Satria Muda Pertamina Jakarta", "Satria Muda Pertamina Jakarta", "Satria Muda Pertamina Jakarta", 
              "Stapac", "Stapac", "Pacific Caesar Surabaya", "Louvre Surabaya"],
    "Country" : ["Indonesia","Indonesia","Indonesia","Indonesia","Indonesia",
                 "Indonesia","Indonesia","Indonesia","Indonesia","Indonesia",],
    "Age" : [24, 21, 23, 26, 20, 21, 25, 23, 21, 24],
    "Height" : [190, 185, 178, 185, 183, 191, 188, 179, 177, 182],
    "Weight" : [90, 86, 83, 75, 80, 85, 90, 87, 80, 85],
    "Avg Points" : [7, 6, 10, 11, 5, 4, 7, 1, 9, 6],
    "Avg Rebounds" : [6, 3, 3, 4, 2, 10, 5, 2, 8, 5]
})
dfRekrut
```

#### Checking & Cleaning Data

```python
enbie.isnull().sum()
```

#### Condition Wanted
- Usia (Age) <= 25 tahun,
- Tinggi badan (Height) >= 180 cm,
- Berat badan (Weight) <= 90 kg,
- Rata-rata point (Average points scored) >= 6, dan
- Rata-rata rebound (Average rebounds grabbed) >= 3.

```python
enbieNeed = enbie[["player_name", "age", "player_height", "player_weight", "pts", "reb"]]
enbieNeed
```

```python
conwtd = enbie[(enbie['age'] <= 25) & (enbie['player_height'] >= 180) & (enbie['player_weight'] <= 90) 
                   & (enbie['pts'] >= 6) & (enbie['reb'] >= 3)]
conwtd.index.values
```

```python
targetList = []
targetIndex = conwtd.index.values

for i in range(len(enbieNeed)):
    if i in targetIndex:
        targetList.append(1)
    else:
        targetList.append(0)

enbieNeed["target"] = pd.DataFrame(targetList)
enbieNeed
```

#### Standarisasi Data

```python
from sklearn.preprocessing import StandardScaler
stds = StandardScaler()
dataSTD = stds.fit_transform(enbieNeed.iloc[:,1:-1]) #exclude target
dataSTD
```

```python
dfUlti = pd.DataFrame(dataSTD)
dfUlti["target"] = enbieNeed["target"]
dfUlti
```

#### Splitting Dataset

```python
x = dfUlti.iloc[:, :-1]
y = np.array(dfUlti["target"])
```

```python
from sklearn.model_selection import train_test_split
xtr, xts, ytr, yts = train_test_split(x, y, test_size= .18)
```

### Hyper-parameter Tuning & Choosing Model

```python
from sklearn.model_selection import RandomizedSearchCV
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
penalty = ["l1", "l2", "elasticnet", "none"]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
max_iter = [1, 10, 100, 1000, 10000]

paramLR = {
    "penalty": penalty,
    "solver": solver,
    "max_iter" : max_iter
}
paramLR

lomo = LogisticRegression()
lomoUlti = RandomizedSearchCV(estimator= lomo, 
                             param_distributions= paramLR, 
                             cv= 5)
lomoUlti.fit(xtr, ytr)
lomoUlti.best_params_
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = round(len(xtr) ** .5)
k
```

```python
n_neighbors = list(range(k-5, k+5))
algo = ["ball_tree", "kd_tree", "brute", "auto"]

paramKN = {
    "n_neighbors": n_neighbors,
    "algorithm": algo
}
paramKN

knmo = KNeighborsClassifier()
knmoUlti = RandomizedSearchCV(estimator= knmo, 
                             param_distributions= paramKN, 
                             cv= 5)
knmoUlti.fit(xtr, ytr)
knmoUlti.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimators = list(range(10, 100, 5))
max_features = ['auto', 'sqrt']
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

paramRF = {'n_estimators': n_estimators,
               'max_features': max_features,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

rfmo = RandomForestClassifier()
rfmoUlti = RandomizedSearchCV(estimator= rfmo, 
                             param_distributions= paramRF, 
                             cv= 5)
rfmoUlti.fit(xtr, ytr)
rfmoUlti.best_params_
```

#### Modeling with Best Parameter

##### Score

```python
yprLM = lomoUlti.predict(xts)
yprKN = knmoUlti.predict(xts)
yprRF = rfmoUlti.predict(xts)
```

```python
print("Score Logistic Regression :", lomoUlti.score(xts, yts))
print("Score KNN :", knmoUlti.score(xts, yts))
print("Score Random Forest :", rfmoUlti.score(xts, yts))
```

##### Balanced Accuracy Score

```python
from sklearn.metrics import balanced_accuracy_score as bas
print("BAS Logistic Regression :", bas(yts, yprLM))
print("BAS KNN :", bas(yts, yprKN))
print("BAS Random Forest :", bas(yts, yprRF))
```

##### Precision Score

```python
from sklearn.metrics import precision_score as ps
print("Precision Logistic Regression :", ps(yts, yprLM))
print("Precision KNN :", ps(yts, yprKN))
print("Precision Random Forest :", ps(yts, yprRF))
```

##### Recall

```python
from sklearn.metrics import recall_score
print("Recall+ Logistic Regression :", recall_score(yts, yprLM))
print("Recall+ KNN :", recall_score(yts, yprKN))
print("Recall+ Random Forest :", recall_score(yts, yprRF))
```

##### F1 Score

```python
from sklearn.metrics import f1_score
print("F1 score Logistic Regression :", f1_score(yts, yprLM))
print("F1 score KNN :", f1_score(yts, yprKN))
print("F1 score Random Forest :", f1_score(yts, yprRF))
```

##### ROC AUC

```python
from sklearn.metrics import roc_auc_score
```

```python
print("AUC score Logistic Regression :", roc_auc_score(yts, lomoUlti.predict(xts)))
print("AUC score KNN :", roc_auc_score(yts, knmoUlti.predict(xts)))
print("AUC score Random Forest :", roc_auc_score(yts, rfmoUlti.predict(xts)))
```

Dari seluruh pengujian, model terbaik yang dapat digunakan adalah rfmoUlti
